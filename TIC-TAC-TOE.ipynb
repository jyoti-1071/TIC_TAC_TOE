```python
import random
import numpy as np
```

```python
def display(board):
    print("\n\n")
    for i in range(3):
        for j in range(3):
            print(board[i][j],end="        ")
        print("\n")   
    print("\n\n")    
```

```python
def initialization( board):
    board=np.array([["-", "-", "-"],
              ["-", "-", "-"],
              ["-", "-", "-"]])    
    return(board)    
```

```python
def checkfill(board):
    for i in board:
        if '-' in i:
            return False
        else:
            return True            
```

```python
def checkwon(board):
    if board[0][0]==board[1][1]==board[2][2]=="O":
        return("O")
    elif  board[0][0]==board[1][1]==board[2][2]=="X":
        return("X")       
    elif board[0][2]==board[1][1]==board[2][0]=="O":
        return("O")   
    elif board[0][2]==board[1][1]==board[2][0]=="X":
        return("X") 
    else:    
        for i in range(3):
            for j in range(3):
                if board[j][0]==board[j][1]==board[j][2]=='X':
                    return("X")
                elif board[j][0]==board[j][1]==board[j][2]=='O':
                    return("O")
                elif board[0][j]==board[1][j]==board[2][j]=="X":
                    return("X")
                elif board[0][j]==board[1][j]==board[2][j]=="O":
                    return("O")  
    return(-1)                 
```

```python
def start_game ():
    board=[]
    game=0
    print("     Board:    \n")
    board=initialization(board)

    choose=['USER1','USER2']
    turn=random.choice(choose)      
    while(game!=1):
        display(board)
        if (checkfill==True):
            print("Tie :(") 
            game=1
            break
        else:
            if turn=='USER1':
                print("USER1")
                print("Enter the position\n")
                m,n=map(int , input().split())
                if board[m][n]=="-":
                    board[m][n]="O"
                else:
                    while board[m][n]!="-":
                        print("Position is cover. Choose any other position\n")
                        m,n=map(int , input().split())
                        if board[m][n]=="-":
                            board[m][n]="O"
                            break
                turn="USER2"
            else:
                print("USER2")
                print("Enter the position\n")
                m,n=map(int , input().split())
                if board[m][n]=="-":
                    board[m][n]="X"
                else:
                    while board[m][n]!="-":
                        print("Position is cover. Choose any other position\n")
                        m,n=map(int , input().split())
                        if board[m][n]=="-":
                            board[m][n]="X"
                            break
                turn="USER1"
            k=checkwon(board)    
            if k=="X":
                    print("X won!!!!!")
                    display(board) 
                    game=1
                    break
            elif k=="O":
                    print("O won!!!!!")
                    display(board) 
                    game=1
                    break
            else:
                continue
```

```python
start_game()
```
